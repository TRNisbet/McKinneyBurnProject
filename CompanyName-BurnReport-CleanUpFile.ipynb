```python
# Import dependencies
import os
import pandas as pd
import datetime
```

```python
# Import ADVANTAGE TIME data into a dataframe
InScope_df = pd.read_csv("Pulte In Scope Hours_.csv")
InScope_df.head()
```

```python
# Pull out only necessary columns
Test_frame = ['Employee', 'Employee Current Title', 'Department Team', 'Job Number', 'Job Description', 'Date', 'Hours']
InScope_df = InScope_df[Test_frame]
InScope_df
```

```python
# Rename the "Department Team" column to "Service Department"
InScope_df = InScope_df.rename(columns={'Department Team': 'Service Department'})

# Rename the "Employee Current Title" column to "Service Description"
InScope_df = InScope_df.rename(columns={'Employee Current Title': 'Service Description'})
InScope_df
```

```python
#Replace Advantage Job Descriptions with Burn Report Formatted Service Descriptions

InScope_df['Service Description'] = InScope_df['Service Description'].replace({
    'Group Client Director': 'Group/Global Client Director',
    'Client Director, Brand Experience': 'Brand Experience Client Director',
    'Account Director': 'Account/Client Director',
    'Project Management Supervisor|Project Manager|Project Manager, Learning & Development': 'Project/Resource Management Mgr/Sup',
    'Art Supervisor|Senior Designer': 'Senior Designer (Art Supe-McK Health)',
    'Content Creator|Creative Apprentice|Senior Content Creator': 'Content sCreator',
    'Art Director|Junior Art Director': 'Art Director',
    'Junior Copywriter': 'Copywriter',
    'Junior Designer': 'Designer'
})

    
```

```python
# Strategy Service Description Consolidation 
InScope_df['Service Description'] = InScope_df['Service Description'].replace({
    'Director of Strategy|Strategy Lead|Director, Consumer & Brand Data': 'Director, Strategy (Strategy Lead-McK Health)',
    'Strategist, Consumer Research': 'Strategist',
    'Group Strategy Director': 'Group Strategy Director, Digital and Social',
    'Strategy Director, Digital and Social|Strategy Director, Social': 'Strategy Director, Digital and Social'
})
```

```python
# Media Service Description Consolidation
InScope_df['Service Description'] = InScope_df['Service Description'].replace({
    'Executive Director, Media': 'Executive Director of Media',
    'Group Media Director, Strategy and Stewardship': 'Group Media Director, Strategy & Stewardship',
    'Group Media Director, Activation and Analytics': 'Group Media Director, Activation & Analytics',
    'Associate Media Director, Strategy and Stewardship': 'Assoc Media Director, Strategy & Stewardship',
    'Associate Media Director, Activation and Analytics': 'Assoc Media Director, Activation & Analytics',
    'Media Supervisor and Media Supervisor, Strategy and Stewardship': 'Media Supervisor, Strategy & Stewardship',
    'Media Supervisor, Activation and Analytics': 'Media Supervisor, Activation & Analytics',
    'Senior Media Planner': 'Sr Media Planner, Strategy & Stewardship/Activation & Analytics',
    'Media Analyst, Activation and Analytics': 'Media Analyst, Activation & Analytics',
    'Media Planner, Activation and Analytics|Media Planner, Strategy and Stewardship': 'Media Planner, Strategy & Stewardship',
    'Junior Programmatic Trader|Programmatic Trader': 'Programmatic Trader, Activation & Analytics',
    'Assistant Media Planner|Assistant Media Planner, Strategy and Stewardship': 'Assistant Media Planner, Strategy & Stewardship',
    'Ad Operations Specialist, Activation & Analytics': 'Ad Operations Specialist'
})
```
